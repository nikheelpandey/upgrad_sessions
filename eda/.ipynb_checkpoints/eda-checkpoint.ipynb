## Dealing with NANs

```python
# remove heavy-weight nan champs
# remove every column that has more than 40% as NaN 

almost_empty_columns = [column for column in df if df[column].count() / len(df) < 0.6]
almost_empty_columns.append('Id')
df_ = df.drop(almost_empty_columns,axis=1)
```

```python
# creating a new df that has only numerical columns
df_num = df_.select_dtypes(include=['float64','int64'])
```
